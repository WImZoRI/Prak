2. Критерий Левене

Может применяться для двух и более групп. Требуется случайность и независимость. Нет требования к нормальности, но мощность снижается при ненормальности

```python
%%R

data_1 <- rnorm(1000, 0, 1)
data_2 <- rgeom(1000, 0.2)
data_3 <- rgeom(1000, 0.21)
data_4 <- rnorm(1000, 0, 1.05)

fact = factor(rep(c(1,2), c(1000,1000)))
```

```python
%%R

print(leveneTest(c(data_1, data_2), fact))
print(leveneTest(c(data_1, data_3), fact))
print(leveneTest(c(data_1, data_4), fact))
```

```python
%%R

print(leveneTest(c(data_2, data_3), fact))
print(leveneTest(c(data_2, data_4), fact))
print(leveneTest(c(data_3, data_4), fact))
```

3. Критерий Бартлетта

# 2. Исследовать корреляционные взаимосвязи в данных с помощью коэффициентов корреляции Пирсона, Спирмена и Кендалла.

### 1. Коэффициент корреляции Пирсона

Этот коэффициент определяет линейную корреляцию между двумя наборами данных с непрерывными значениями. Он чувствителен к выбросам, требует независимость наблюдений и желательна нормальность

Для его применения мы должны предположить, что данные зависят друг от друга линейно. Для этого построим scatter plot

```python
%%R
columns_for_print <- c('Respiratory_Rate', 'Oxygen_Saturation', 'Systolic_BP', 'Heart_Rate', 'Temperature')
count <- 0
par(mfrow = c(3, 4), mar = c(4, 4, 2, 1))

for (i in 1:(length(columns_for_print) - 1)) {
    for (j in (i + 1):length(columns_for_print)) {
        count <- count + 1
        first <- columns_for_print[i]
        second <- columns_for_print[j]
        plot(ds[[first]], ds[[second]], col = "skyblue", xlab = first, ylab = second)
    }
}
```

```python
%%R

for (i in 1:(length(columns_for_print) - 1)) {
    for (j in (i + 1):length(columns_for_print)) {
        first <- columns_for_print[i]
        second <- columns_for_print[j]
        data_1 <- ds[[first]]
        data_2 <- ds[[second]]
        cor <- cor.test(data_1, data_2, method = "pearson")
        print(paste(first, second, "cor =", cor$estimate,  "p-value =", cor$p.value))
    }
}
```

Искусственный пример

```python
%%R

x <- rnorm(1000)
y <- 2 * x + 5 + rnorm(1000) * 0.01
plot(x, y)
cor <- cor.test(x, y, method = "pearson")
print(paste("cor =", cor$estimate,  "p-value =", cor$p.value))
```

```python
%%R

x <- rnorm(1000)
y <- 2 * x ^ 2 + 5 + rnorm(1000) * 0.01
plot(x, y)
cor <- cor.test(x, y, method = "pearson")
print(paste("cor =", cor$estimate,  "p-value =", cor$p.value))
```

### 2. Коэффициент корреляции Спирмена

Коэффициент корреляции Спирмена определяет монотонную корреляцию между данными.

Данные должны быть непрерывными и независимыми, имеется предположение о монотонности данных, не так чувствителен к выбросам 

```python
%%R

print(cor.test(ds[['On_Oxygen']], ds[['Heart_Rate']], method = "pearson"))
print(cor.test(ds[['Oxygen_Saturation']], ds[['Heart_Rate']], method = "pearson"))
```

```python
%%R

print(cor.test(ds[['Respiratory_Rate']], ds[['Heart_Rate']], method = "pearson"))
print(cor.test(ds[['O2_Scale']], ds[['Heart_Rate']], method = "pearson"))
```

```python
%%R

print(cor.test(ds[['Systolic_BP']], ds[['Heart_Rate']], method = "pearson"))
print(cor.test(ds[['Temperature']], ds[['Heart_Rate']], method = "pearson"))
```

```python
%%R

x <- rnorm(1000)
y <- 2 * x + 5 + rnorm(1000) * 0.01
plot(x, y)
cor <- cor.test(x, y, method = "spearman")
print(paste("cor =", cor$estimate,  "p-value =", cor$p.value))
```

```python
%%R

x <- rnorm(1000)
y <- 2 * x ^ 2 + 5 + rnorm(1000) * 0.01
plot(x, y)
cor <- cor.test(x, y, method = "spearman")
print(paste("cor =", cor$estimate,  "p-value =", cor$p.value))
```

### 3. Коэффициент корреляции Кендалла

Коэффициент корреляции Кендалла определяет монотонную корреляцию между данными.

Требует непрерывные, независимые данные

```python
%%R

for (i in 1:(length(columns_for_print) - 1)) {
    for (j in (i + 1):length(columns_for_print)) {
        first <- columns_for_print[i]
        second <- columns_for_print[j]
        data_1 <- ds[[first]]
        data_2 <- ds[[second]]
        cor <- cor.test(data_1, data_2, method = "kendall")
        print(paste(first, second, "cor =", cor$estimate,  "p-value =", cor$p.value))
    }
}
```

```python
%%R

x <- rnorm(1000)
y <- 2 * x + 5 + rnorm(1000) * 0.01
plot(x, y)
cor <- cor.test(x, y, method = "kendall")
print(paste("cor =", cor$estimate,  "p-value =", cor$p.value))
```

### 3. Тест МакНемара

Применяется для парных данных, например, данные о пациенте до применения препарата и после. Тест позволяет понять, изменились ли результаты

В нашем датасете нет реальных парных данных, поэтому создадим искусственно

```python
%%R

before <- sample(c(0, 1), 1000, replace=TRUE, prob=c(0.7, 0.3))
after_1 <- sample(c(0, 1), 1000, replace=TRUE, prob=c(0.6, 0.4))
after_2 <- sample(c(0, 1), 1000, replace=TRUE, prob=c(0.69, 0.31))
```

```python
%%R
print(mcnemar.test(table(before, after_1)))
print(mcnemar.test(table(before, after_2)))
```

# 5. Исследовать зависимости в данных с помощью дисперсионного анализа.

Дисперсионный анализ применяется для сравнения средних значений переменной в разных группах. Для применения остатки должны иметь около нормальное распределение, но при достаточно больших объемах данных и можно применять, если нет большого количества выбросов. Также дисперсии должны быть примерно равны

```python
%%R
g1 <- rnorm(100,0,1)
g2 <- rnorm(100,5,1)
g3 <- rnorm(100,10,1)
data <- data.frame(
  value = c(group1, group2, group3),
  group = factor(rep(c('A', 'B', 'C'), each = 100))
)
result <- aov(value ~ group, data = data)
anova_summary <- summary(result)
cat("F-statistic:", anova_summary[[1]]$`F value`[1], "\n")
cat("p-value:", anova_summary[[1]]$`Pr(>F)`[1], "\n\n")
```

# 6. Подогнать регрессионные модели (в том числе, нелинейные) к данным, а также оценить качество подобной аппроксимации.

Воспользуемся полиномиальной логистической регрессией для предсказания Risk_Level. Для оценки модели будем использовать accuracy_score

```python
%%R
ds <- read.csv('Health_Risk_Dataset.csv')
ds$Patient_ID <- NULL
ds_to_reg <- as.data.frame(lapply(ds, function(x) as.numeric(factor(x))))
```

```python
%%R
X <- ds_to_reg[, names(ds_to_reg) != "Risk_Level"]
y <- ds_to_reg$Risk_Level
train_index <- sample(1:nrow(X), 0.75 * nrow(X))
X_train <- X[train_index, ]
X_valid <- X[-train_index, ]
y_train <- y[train_index]
y_valid <- y[-train_index]
X_train_scaled <- scale(X_train)
X_valid_scaled <- scale(X_valid, center = attr(X_train_scaled, "scaled:center"), scale = attr(X_train_scaled, "scaled:scale"))
model <- multinom(as.factor(y_train) ~ ., data=as.data.frame(scale(X_train)), maxit=2000, MaxNWts=10000, decay=1/100)
y_pred <- predict(model, as.data.frame(X_valid_scaled))
mean(y_pred == as.factor(y_valid))
```

```python
%%R
library(glmnet)
X <- as.matrix(ds_to_reg[, names(ds_to_reg) != "Risk_Level"])
y <- as.numeric(factor(ds_to_reg$Risk_Level)) - 1
idx <- sample(1:nrow(X), 0.75 * nrow(X))

X_train <- scale(X[idx, ])
X_test <- scale(X[-idx, ], 
                center = attr(X_train, "scaled:center"),
                scale = attr(X_train, "scaled:scale"))
y_train <- y[idx]
y_test <- y[-idx]

model <- glmnet(X_train, y_train, alpha=0, lambda=0.01, family="multinomial")
pred <- predict(model, X_test, type="response")
pred_classes <- apply(pred[, , 1], 1, which.max) - 1

cat("Accuracy:", round(mean(pred_classes == y_test), 4), "\n")
```
